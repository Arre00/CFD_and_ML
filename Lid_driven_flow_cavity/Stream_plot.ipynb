```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.externals import joblib
import pickle

if IN_COLAB:
    U = np.load('/content/drive/MyDrive/Lid_Driven_Cavity/Solution_5_U.npy')
    V = np.load('/content/drive/MyDrive/Lid_Driven_Cavity/Solution_5_V.npy')
    p = np.load('/content/drive/MyDrive/Lid_Driven_Cavity/Solution_5_p.npy')


else:
    U = np.load('Solution_5_U.npy')
    V = np.load('Solution_5_V.npy')
    p = np.load('Solution_5_p.npy')

M = U.shape[1] - 2

Dx = Dy = 1 / M

# M x M cells + Ghost cells on either boundaries
x = np.linspace(-Dx/2, 1 + Dx/2, M+2)
y = np.linspace(-Dy/2, 1 + Dy/2, M+2)

X, Y = np.meshgrid(x, y, indexing='ij')
plt.figure(figsize=[10, 10])

plt.streamplot(X.T, Y.T, U[-1].T, V[-1].T, 4, arrowstyle='-')
plt.xlim([0,1])
plt.ylim([0,1])

plt.xlabel('x')
plt.ylabel('y')

plt.title(r'Streamplot')
plt.show()
```
